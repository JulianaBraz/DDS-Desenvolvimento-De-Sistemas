# **A metodologia Cascata**

 --> É um modelo de gestão de projetos sequencial e linear, onde cada fase do projeto é concluída antes de se iniciar a próxima, fluindo como uma cachoeira. As etapas típicas incluem requisitos, design, implementação, verificação (testes) e manutenção. É um método rígido, com pouca flexibilidade para mudanças após a conclusão de uma fase, tornando-o ideal para projetos com requisitos estáveis e bem definidos, mas desaconselhado para projetos com incertezas ou necessidade de adaptação.

 **Como funciona:**
1. Sequência linear:
   > As fases são executadas numa ordem estrita, sem sobreposição.
2. Conclusão de fase:
   > Cada etapa deve ser totalmente concluída e aprovada antes que a próxima possa começar.
3. Planeamento detalhado:
   > O planeamento é a base do projeto, com um grande foco em documentação e na definição prévia de todos os detalhes.
4. Fluxo rígido:
   > Não há flexibilidade para voltar a etapas anteriores e corrigir erros, o que pode exigir a refação de grande parte do projet.

 **Fases:**

* Requisitos: *Levantamento e documentação detalhada dos requisitos do projeto.*
* Design: Criação do plano e da arquitetura da solução.
* Implementação: *Desenvolvimento e codificação do produto.*
* Verificação / Teste: *Teste do produto para garantir que atende aos requisitos e não contém erros.*
* Manutenção: *Correção de bugs e atualizações após a entrega do produto.*



 **Vantagens:**
1.   Estrutura e organização:
*Proporciona um processo claro, previsível e bem definido.*
2.   Ideal para projetos estáveis:
*Funciona bem quando os requisitos são claros, fixos e não há previsão de mudanças.*
3.   Orçamento e prazo definidos:
*Permite a estipulação de custos e prazos fixos desde o início.*


**Desvantagens:**
1.   Pouca flexibilidade:
*Inadequado para projetos com requisitos voláteis, pois é muito difícil fazer alterações.*
2.   Risco de refazer o trabalho:
*Erros ou falhas em fases posteriores podem exigir a refação de etapas anteriores.*
3.   Ociosidade da equipe:
*A dependência sequencial pode levar a ociosidade de alguns membros da equipe.*
4.   Entrega final tardia:
*A entrega do produto só ocorre no final do ciclo de vida do projet*

# **TDD - Test-Driven Development**

  **Desenvolvimento Orientado a Testes**
   
  --> É uma metodologia de desenvolvimento de software onde os testes são escritos antes do código que será testado. O processo envolve um ciclo iterativo de escrever um teste, fazer com que ele falhe (porque o código ainda não existe), escrever o código mínimo para passar no teste e, finalmente, refatorar o código. Ele opera no ciclo RED (escreva um teste que falha), GREEN (faça o teste passar com o mínimo de código) e REFACTOR (melhore o código mantendo os testes passando). Este processo garante que o código seja testável, melhora o design e atua como documentação viva.


   **Como funciona o TDD:**
1. Escrever um teste:
O desenvolvedor cria um teste automatizado que define o comportamento esperado de uma parte do código.
2. Executar e Falhar:
O teste é executado e, como o código correspondente ainda não existe, ele falha.
3. Escrever o código:
O desenvolvedor escreve a implementação mais simples possível para fazer o teste passar.
4. Refatorar:
O código é revisado e melhorado, sem alterar seu comportamento, para torná-lo mais limpo e eficiente.
5. Repetir:
O ciclo é repetido para novas funcionalidades ou alterações no código.

**Benefícios do TDD:**

*   Código mais limpo e com menos erros: Ao escrever testes primeiro, os desenvolvedores são forçados a entender bem o problema e a escrever código mais simples e modularizado, o que reduz a probabilidade de erros.

*  Melhora a documentação: Os testes servem como documentação do código, mostrando como ele deve funcionar.


*   Facilita a manutenção e refatoração: O TDD torna mais fácil a identificação e correção de problemas, além de facilitar a refatoração do código sem introduzir novos erros.

*   Aumenta a confiança no código: Como o código é testado continuamente, a confiança no seu funcionamento aumenta.

*   Ciclo de desenvolvimento mais curto: Ao identificar e corrigir problemas mais cedo, o TDD pode reduzir o tempo total de desenvolvimento.

# **XP - Extreme Programming**

--> É uma metodologia ágil abrangente, também criada por Kent Beck, que visa melhorar a qualidade do software e a capacidade de resposta às mudanças nos requisitos do cliente. Ela é caracterizada por ciclos de desenvolvimento curtos, planejamento incremental e feedback contínuo. Os cinco valores fundamentais do XP são:

• Comunicação: Constante e eficaz.

• Simplicidade: Fazer a coisa mais simples que funcione.

• Feedback: Rápido e contínuo.

• Coragem: Para fazer mudanças.

• Respeito: Mútuo entre a equipe.

--> O diferencial do XP em relação a outras metodologias ágeis é que ele oferece práticas técnicas específicas para aprimorar a qualidade do código e a produtividade da equipe

# **Sinergia TDD e XP**

A integração de TDD e XP é uma das principais razões para o sucesso e a robustez dessas abordagens. O TDD é uma das práticas fundamentais do XP, agindo como um pilar técnico dentro da estrutura mais ampla do XP.

**• Como TDD fortalece o XP:**

    ◦ Suporta o design simples ao evitar código desnecessário.
    ◦ Facilita a refatoração com segurança.
    ◦ Promove integração contínua com testes automatizados.

**• Como XP potencializa o TDD:**

    ◦ A programação em par melhora a qualidade dos testes.
    ◦ A prática de ter o cliente no local ajuda a definir testes mais relevantes e a obter feedback rápido.

• Valores Compartilhados: Ambas as abordagens compartilham valores como simplicidade, feedback, coragem e a ideia de testes como comunicação, alinhando-se perfeitamente com os princípios ágeis.

--> Resultados da Adoção Conjunta A combinação de TDD e XP leva a resultados significativos que reforçam os objetivos das metodologias ágeis:

• Maior qualidade do código, com menos bugs e problemas em produção.

• Melhor design e arquitetura de software.

• Maior agilidade e capacidade de adaptação a mudanças.

• Produtividade da equipe aumentada a médio e longo prazo, com menos tempo gasto em depuração e retrabalho.

• Melhor colaboração e compartilhamento de conhecimento, especialmente através da programação em par e testes como documentação.

• Maior satisfação dos stakeholders com entregas mais previsíveis e confiáveis.
Em essência, a "Introdução às Metodologias Ágeis" estabelece o terreno, definindo os valores e princípios que guiam o desenvolvimento.

     > TDD e XP, criados pelo mesmo indivíduo, são exemplares dessas metodologias.

     resultando em software de alta qualidade e equipes mais eficazes.

# **DIFERENÇA ENTRE:**

# - METODOLOGIA CASCATA VS METODOLOGIA ÁGIL

*As metodologias de desenvolvimento de software, como cascata e ágil, diferem principalmente na forma como gerenciam o ciclo de vida do projeto e a adaptabilidade a mudanças. A metodologia cascata é um modelo sequencial e linear, onde cada fase do projeto (requisitos, design, implementação, testes, etc.) é concluída antes de passar para a próxima. Já a metodologia ágil é iterativa e incremental, permitindo adaptações e mudanças em cada iteração ou "sprint" do projeto.*

# - QA VS SALA DE GUERRA
*A principal diferença é que QA refere-se a uma função e área de atuação, focada em garantir a qualidade de software através de testes e análise, enquanto uma Sala de Guerra é um método, um espaço ou uma estrutura de trabalho onde pessoas de diversas áreas se reúnem para focar intensamente num projeto, resolver um problema urgente ou alcançar uma meta específica, com forte ênfase em colaboração e tomada rápida de decisões.*

# **MVP - Minimum Viable Product**
- Produto Mínimo Viável:

 > É uma versão simplificada de um produto que contém apenas os recursos essenciais para resolver um problema inicial, permitindo que a equipe colete feedback valioso dos usuários reais para validar a ideia e guiar o desenvolvimento futuro com o mínimo de esforço e recursos. Essa abordagem visa reduzir o desperdício de tempo e dinheiro, acelerar o lançamento e adaptar a estratégia com dados concretos antes de um investimento maior, especialmente em mercados voláteis

- Como Funciona:
1. Identificação da Hipótese Central:
*Define-se a principal necessidade que o produto pretende resolver.*
2. Desenvolvimento do Escopo Essencial:
*A equipe cria uma versão do produto com as funcionalidades mais básicas e críticas.*
3. Lançamento para Usuários Iniciais:
*O MVP é disponibilizado para um grupo de usuários, que podem ser clientes em potencial.*
4. Coleta de Feedback:
*Os usuários testam o produto e fornecem informações sobre o que funciona, o que pode ser melhorado e o que falta.*
5. Iteração e Aprimoramento:
*Com base no feedback, a equipe toma decisões sobre quais novas funcionalidades desenvolver, quais manter e quais remover, ajustando o produto de forma contínua*


- Benefícios do MVP:
>Agilidade: *Reduz o tempo de lançamento e permite uma adaptação mais rápida às mudanças do mercado.*

 >Inovação Eficaz: *Acelera o processo de inovação ao validar ideias de forma segura.*

 >Foco no Cliente: *Garante que o produto final atenda às necessidades reais dos usuários.*

 >Redução de Custos: *Evita investimentos desnecessários em funcionalidades que não agregam valo*

# **MVC - MODEL VIEW CONTROLLER**

-  É um padrão de arquitetura de software responsável por contribuir na otimização da velocidade entre as requisições feitas pelo comando dos usuários.
- MVC é dividida em três componentes essenciais: Model, Controller e View:
   >MODEL: *Essa classe também é conhecida como Business Object Model (objeto modelo de negócio). Sua responsabilidade é gerenciar e controlar a forma como os dados se comportam por meio das funções, lógica e regras de negócios estabelecidas. Ele é o detentor dos dados que recebe as informações do Controller, válida se ela está correta ou não e envia a resposta mais adequada.*

   >CONTROLLER: *A camada de controle é responsável por intermediar as requisições enviadas pelo View com as respostas fornecidas pelo Model, processando os dados que o usuário informou e repassando para outras camadas.
   Numa analogia bem simplista, o controller  operaria como o ‘’maestro de uma orquestra’’  que permite a comunicação entre o detentor dos dados e a pessoa com vários questionamentos no MVC.*

   >VIEW: *Essa camada é responsável por apresentar as informações de forma visual ao usuário. Em seu desenvolvimento devem ser aplicados apenas recursos ligados a aparência como mensagens, botões ou telas.
   Seguindo nosso processo de comparação o View está na linha de frente da comunicação com usuário e é responsável transmitir questionamentos ao controller e entregar as respostas obtidas ao usuário. É a parte da interface que se comunica, disponibilizando e capturando todas as informação do usuário.*


mvc.png

Tudo começa com a interação do usuário na camada View. A partir daí o controlador pega essa informações e envia para o Model que fica responsável por avaliar aqueles dados e transmitir uma resposta.
O controlador recebe essas respostas e envia uma notificação de validação daquela informação para a camada visão, fazendo com a mesma apresente o resultado de maneira gráfica e visual.

- Exemplos: MODEL VIEW CONTROLLER

Model: classes de banco de dados, que vão gerar tabelas, carnadialidade

View: TELA: html, Swing, Java, CSS, JFrame, "parte da frente"

Controller: configura as rotas(encaminhamento: sub url(sub rota)). ex: /drive/

# BENEFÍCIOS:

*Além disso, um software precisa ter estabilidade no processo de comunicação entre seus elementos de maneira dinâmica para que a experiência do usuário não seja prejudicada.*

# **SOLID**
SOLID é um acrónimo que representa um conjunto de cinco princípios de design de software para programação orientada a objetos, que visam criar sistemas mais flexíveis, escaláveis e fáceis de manter e evoluir.

Os 5 Princípios são:


1.   **S - Single Responsibility Principle (Princípio da Responsabilidade Única):**
*Uma classe ou módulo deve ter apenas uma razão para ser modificada, ou seja, deve ter um único propósito ou responsabilidade bem definida.*

2.   **O - Open-Closed Principle (Princípio Aberto-Fechado):**
*O software (classes, módulos, funções) deve ser aberto para extensão, mas fechado para modificação. Novas funcionalidades devem ser adicionadas sem alterar o código já existente.*

3.   **L - Liskov Substitution Principle (Princípio da Substituição de Liskov):**
*Objetos de uma classe base devem poder ser substituídos por objetos de uma classe derivada sem que o programa seja afetado.*

4.   **I - Interface Segregation Principle (Princípio da Segregação da Interface):**
*Os clientes não devem ser forçados a implementar interfaces que não utilizam. É melhor ter interfaces menores e específicas para cada tipo de cliente.*

5.  **D - Dependency Inversion Principle (Princípio da Inversão da Dependência):**
*Módulos de alto nível não devem depender de módulos de baixo nível. Ambos devem depender de abstrações (como interfaces).*

* Manutenibilidade: *O código torna-se mais fácil de entender e manter.*
* Reutilização: *Facilita a reutilização de código em diferentes partes do projeto ou em outros projetos.*

* Escalabilidade: *O software pode ser expandido e adaptado de forma mais simples a novas necessidades.*
